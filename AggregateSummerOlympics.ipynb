# Count total medals for each event(year,sport,discipline,event,event_gender)

```python
summer['Event_Medals'] = summer.groupby(['Year','Sport','Discipline','Event','Event_Gender']).Medal.transform('count')
```

# Any event with more than 5 medals will be designated as team events

```python
summer['Team'] = False
```

```python
summer
```

```python
summer.loc[summer.Event_Medals > 5,'Team'] = True
```

```python
single = summer[summer.Team == False].copy()
teams = summer[summer.Team == True].copy()
```

```python
teams = teams.drop_duplicates(subset=['Year','Sport','Discipline','Country','Event','Medal','Event_Gender'])
```

```python
overall = pd.concat([single,teams],axis=0)
```
